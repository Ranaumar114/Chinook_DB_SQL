```python
 q = """
    SELECT c.first_name || " " || c.last_name Customer_Name , COUNT(inv.invoice_id) No_OF_Purchase,
    SUM(inv.total) Total_Spend FROM customer c
    LEFT JOIN invoice inv
    ON
    c.customer_id = inv.customer_id
    GROUP BY c.customer_id
    ORDER BY 3 DESC
    
    
"""
pd.read_sql_query(q , conn)
```

```python
# MAKE NEW COLUMNS with case and end use
q = """
    SELECT c.first_name || " " || c.last_name Customer_Name , COUNT(inv.invoice_id) No_OF_Purchase,
    SUM(inv.total) Total_Spend,
    CASE
        WHEN SUM(inv.total) < 40 then 'Small Spender'
        WHEN SUM(inv.total) > 100 then 'Big Spender'
        ELSE "Regular Spender"
    END AS Customer_Category
    FROM customer c
    LEFT JOIN invoice inv
    ON
    c.customer_id = inv.customer_id
    GROUP BY c.customer_id
    ORDER BY 3 DESC
    
    
"""
pd.read_sql_query(q , conn)
```

```python
# MAKE NEW COLUMNS with case and end use, Now find Counts of Category
q = """
    WITH cust_category as (
    SELECT c.first_name || " " || c.last_name Customer_Name , COUNT(inv.invoice_id) No_OF_Purchase,
    SUM(inv.total) Total_Spend,
    CASE
        WHEN SUM(inv.total) < 40 then 'Small Spender'
        WHEN SUM(inv.total) > 100 then 'Big Spender'
        ELSE "Regular Spender"
    END AS Customer_Category
    FROM customer c
    LEFT JOIN invoice inv
    ON
    c.customer_id = inv.customer_id
    GROUP BY c.customer_id
    ORDER BY 3 DESC
    )
    SELECT Customer_Category , COUNT(Customer_Category) Value_Counts FROM cust_category
    GROUP BY 1
    
"""
pd.read_sql_query(q , conn)
```
